```python
import pandas as pd
```

```python
cols = ['Week 1 Day 1', 'Week 1 Day 2', 'Week 1 Day 3', 
        'Week 1 Day 4', 'Week 1 Day 5', 'Week 2 Day 1']

arrival_time = pd.read_csv('data.csv')
arrival_time.columns = cols
```

```python
arrival_time
```

```python
#check the type of object
type(arrival_time['Week 1 Day 1'][0])
```

```python
#convert string to date_time
arrivals_in_datetime = arrival_time.loc[:,cols].apply(pd.to_datetime, errors = 'coerce')

arrivals_in_datetime
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.dates as mdates #matplotlib date formating
```

```python
#Create a list of arrival times for the second day
day_two_times = list(arrival_time['Week 1 Day 2'])
day_two_times
```

```python
# create a figure and plot a histogram of first day arrival times
fig, ax = plt.subplots(figsize=(10,6))

plt.title('Day 2 Arrival Times')
plt.xlabel('Arrival Time')
plt.ylabel('Frequency')


plt.hist(day_two_times, bins = 50)
plt.show()
```
