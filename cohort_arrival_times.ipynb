Objective:

You want to understand when your cohort typically arrives on campus.

Create a visualization of the data including one histogram



```python
import pandas as pd
```

```python
arrival_time = pd.read_csv('data.csv', names = ['Week 1 Day 1', 'Week 1 Day 2', 'Week 1 Day 3', 
                                                 'Week 1 Day 4', 'Week 1 Day 5', 'Week 2 Day 1'])
```

```python
arrival_time
```

```python
type(arrival_time['Week 1 Day 1'][0])
```

```python
# replace absent values with NULL values
for day in arrival_time:
    arrival_time.loc[arrival_time[day] == 'absent', day] = None

arrival_time
```

```python
#convert string to date_time
arrival_time.applymap(pd.Timestamp)
```

```python
import matplotlib.pyplot as plt
```

```python
#Create a list of arrival times for the second day
x = arrival_time['Week 1 Day 2']
```

```python
# create a figure and plot a histogram of first day arrival times
fig = plt.figure(figsize=(8,6))

ax = fig.add_axes()

plt.title('Day 2 Arrival Times')
plt.xlabel('Arrival Time')
plt.ylabel('Frequency')

plt.hist(x, bins = 50)
plt.show()
```
